# **Diabetes Prediction**

## **Import the Dependencies**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

## **Data Collection and Processing**

```python
# Loading the dataset to pandas Dataframe
diabetes_df = pd.read_csv('diabetes.csv')
# Displaying the top 5 records
diabetes_df.head()
```

```python
# Total number of rows and columns
diabetes_df.shape
```

```python
# getting the total count for 0 and 1
diabetes_df['Outcome'].value_counts()
```



* 0 - Non-diabetic
* 1 - Diabetic




```python
# Checking whether any missing values are present
if (diabetes_df.isnull().sum().any()>0):
  print('Yes, there are missing values')
else:
  print('No, there are no missing values')
```

```python
# Checking the data type
diabetes_df.info()
```

```python
# Statistical information
diabetes_df.describe()
```

```python
# Grouping the data based on 0 and 1
# Finding the mean value of it
diabetes_df.groupby('Outcome').mean()
```

```python
# Splitting the data into Independent and Dependent Features
X = diabetes_df.iloc[:,:8]
y = diabetes_df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

## **Data Scaling**

```python
# object of StandardScaler
scaler = StandardScaler()
```

```python
# for getting data in the same range
scaled_data = scaler.fit_transform(X)
```

```python
# displaying scaled data
print(scaled_data)
```

```python
X = scaled_data
```

```python
print(X)
print(y)
```

## **Splitting the data into Train and Test**

```python
# Use stratify to ensure balanced class distribution
# It's particularly important for imbalanced datasets
# Without stratify, your train and test sets may have very different distributions, leading to biased results

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
# Original data, Training data, Test data
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

## **Model Training**
### **Training the SVM Model**

```python
# using Support Vector Classifier model
classifier = svm.SVC(kernel='linear')
```

```python
# training the model
classifier.fit(X_train, y_train)
```

## **Model Evaluation**
### **Evaluating the SVM Model**

```python
# Finding accuracy score of training data
X_train_pred = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, y_train)
print('Accuracy Score of Training Data:',training_data_accuracy)
```

```python
# Finding accuracy score of test data
X_test_pred = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, y_test)
print('Accuracy Score of Test Data:', test_data_accuracy)
```

## **Predictive System**

```python
# Input data
input_data = (9,171,110,24,240,45.4,0.721,54)

# Feature names (exclude 'Outcome')
feature_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Convert input to Pandas DataFrame
input_df = pd.DataFrame([input_data], columns=feature_names)

# Standardize the input data
std_data = scaler.transform(input_df)

# Prediction
prediction = classifier.predict(std_data)

# Output the result
if prediction[0] == 0:
    print('Non-diabetic')
else:
    print('Diabetic')
```
